```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
y
```

```python
x
```

```python
plt.plot(x,y,'g')
plt.xlabel('Time')
plt.ylabel('Points')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
#OO
fig = plt.figure()

axes = fig.add_axes([.1, .1, .8, .8])
axes2 = fig.add_axes([.2,.5,.4,.3])
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes2.plot(y,x)
```

```python
fig = plt.figure(figsize=(3,2))
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2, label='X^2')
axes.plot(x,x**3, label='X^3')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.legend()
```

```python
fig.savefig('myfig.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',linewidth=3,marker='o')
```

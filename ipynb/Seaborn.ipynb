```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.histplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips,hue='sex')
```
